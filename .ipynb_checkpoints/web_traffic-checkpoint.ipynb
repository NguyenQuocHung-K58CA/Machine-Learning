```python
import scipy as sp
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

data = sp.genfromtxt("Data/ch01/data/web_traffic.tsv", delimiter="\t")

print data.shape
des = [1, 2, 3, 10, 50]
linestyles = ['-', '--', ':', '-.', '-']
colors = ['g', 'k', 'b', 'm', 'r']
errors = []
f = []

x = data[:, 0]
y = data[:, 1]

x = x[~sp.isnan(y)]
y = y[~sp.isnan(y)]
```

```python

plt.scatter(x, y)
plt.title("Web trafic over the last month")
plt.xlabel("Time")
plt.ylabel("Hits/Hours")
plt.xticks([w*7*24 for w in range(10)], ['week %i'%w for w in range(10)])
plt.autoscale(tight=True)
plt.grid()
plt.show()
```

```python
def error(f, x, y):
    return sp.sum((f(x)-y)**2)

for i, d in enumerate(des):
    fp, res, rank, sv, rcond = sp.polyfit(x, y, d, full=True)
    tmp = sp.poly1d(fp)
    f.append(tmp)
    errors.append(error(tmp, x, y))
    
    
fx = sp.linspace(0, x[-1], 1000)
for i, ff in enumerate(f):
    plt.plot(fx, ff(fx), linewidth = 2, linestyle = linestyles[i], c = colors[i])

plt.legend([("d=%i" %ff.order) for ff in f], loc="upper left")

for i, e in enumerate(errors):
    print "Error d=%i : %f" %(des[i], e)

inflection = int(3.5*7*24)
xl = x[:inflection]
yl = y[:inflection]

xr = x[inflection:]
yr = y[inflection:]


fl = sp.poly1d(sp.polyfit(xl, yl, 1))
fr = sp.poly1d(sp.polyfit(xr, yr, 1))

fl_error = error(fl, xl, yl)
fr_error = error(fr, xr, yr)

print fr_error
```

```python
split_idx = int(0.3 * len(xr))
shuffled = sp.random.permutation(len(xr))

test = sorted(shuffled[:split_idx])
train = sorted(shuffled[split_idx:])

x_test = xr[test]
y_test = yr[test]
x_train = xr[train]
y_train = yr[train]

fb1 = sp.poly1d(sp.polyfit(x_train, y_train, 1))
fb2 = sp.poly1d(sp.polyfit(x_train, y_train, 2))

fb1_error = error(fb1, x_test, y_test)
fb2_error = error(fb2, x_test, y_test)

print fb1_error
print fb2_error

print fsolve(fb2-100000, 800)/(7*24)
```
