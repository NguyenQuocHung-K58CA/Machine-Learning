```python
from matplotlib import pyplot as plt
import scipy as sp
import numpy as np
from sklearn.datasets import load_iris

data = load_iris()

features = data['data']

feature_names = data['feature_names']

target = data['target']
target_names = data['target_names']
setosa = target_names=='setosa'

plength = features[:, 2]


for f1 in range(4):
    for f2 in range(f1+1, 4):
        for t, marker, c in zip(xrange(3), ">ox", "rgb"):
            plt.scatter(features[target == t, f1],
                        features[target == t, f2],
                        marker = marker,
                        c = c)
            plt.xlabel(feature_names[f1])
            plt.ylabel(feature_names[f2])
        plt.show()
```
